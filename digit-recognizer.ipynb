```python
train = pd.read_csv("../input/digit-recognizer/train.csv")
train
```

```python
X = train.iloc[:, 1:]
Y = train.iloc[:, 0]
```

# Splitting the dataset into training and validation dataset

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_Y, val_Y = train_test_split(X, Y, test_size=0.2, shuffle=True, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score
```

## DecisionTreeClassifier

```python
dtc = DecisionTreeClassifier()
dtc_scores = cross_val_score(dtc, train_X, train_Y, cv=5, scoring="f1_weighted")
print(dtc_scores, dtc_scores.mean())
```

## RandomForestClassifier

```python
rfc = RandomForestClassifier(n_estimators=100)
rfc_scores = cross_val_score(rfc, train_X, train_Y, cv=5, scoring="f1_weighted")
print(rfc_scores, rfc_scores.mean())
```

```python
rfc = RandomForestClassifier(n_estimators=200)
rfc_scores = cross_val_score(rfc, train_X, train_Y, cv=5, scoring="f1_weighted")
print(rfc_scores, rfc_scores.mean())
```

## KNeighborsClassifier

```python
knc = KNeighborsClassifier(n_neighbors=5)
knc_scores = cross_val_score(knc, train_X, train_Y, cv=5, scoring="f1_weighted")
print(knc_scores, knc_scores.mean())
```

```python
knc = KNeighborsClassifier(n_neighbors=10)
knc_scores = cross_val_score(knc, train_X, train_Y, cv=5, scoring="f1_weighted")
print(knc_scores, knc_scores.mean())
```

```python
knc = KNeighborsClassifier(n_neighbors=6)
knc_scores = cross_val_score(knc, train_X, train_Y, cv=5, scoring="f1_weighted")
print(knc_scores, knc_scores.mean())
```

```python
knc = KNeighborsClassifier(n_neighbors=4)
knc_scores = cross_val_score(knc, train_X, train_Y, cv=5, scoring="f1_weighted")
print(knc_scores, knc_scores.mean())
```

## XGBoost

```python
xgb = XGBClassifier(n_estimators=2, objective="multi:softmax")
xgb_scores = cross_val_score(xgb, train_X, train_Y, cv=5, scoring="f1_weighted")
print(xgb_scores, xgb_scores.mean())
```

```python
xgb = XGBClassifier(n_estimators=10, objective="multi:softmax")
xgb_scores = cross_val_score(xgb, train_X, train_Y, cv=5, scoring="f1_weighted")
print(xgb_scores, xgb_scores.mean())
```

```python
xgb = XGBClassifier(n_estimators=15, objective="multi:softmax")
xgb_scores = cross_val_score(xgb, train_X, train_Y, cv=5, scoring="f1_weighted")
print(xgb_scores, xgb_scores.mean())
```

```python
xgb = XGBClassifier(n_estimators=20, objective="multi:softmax")
xgb_scores = cross_val_score(xgb, train_X, train_Y, cv=5, scoring="f1_weighted")
print(xgb_scores, xgb_scores.mean())
```

```python
xgb = XGBClassifier(n_estimators=50, objective="multi:softmax", learning_rate=1)
xgb_scores = cross_val_score(xgb, train_X, train_Y, cv=5, scoring="f1_weighted")
print(xgb_scores, xgb_scores.mean())
```

## GaussianNB

```python
gnb = GaussianNB()
gnb_scores = cross_val_score(gnb, train_X, train_Y, cv=5, scoring="f1_weighted")
print(gnb_scores, gnb_scores.mean())
```
